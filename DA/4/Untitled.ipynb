```python
import glob
import numpy as np
import pandas as pd
```

```python
path="./"
data = []
all_files = glob.glob(path+"/*.csv")
for file in all_files:
    frame = pd.read_csv(file,index_col=None,header=0)
    data.append(frame)
df = pd.concat(data,axis=0,ignore_index=True)
```

```python
df.head()
```

```python
df = df.drop(labels=['Start date','End date','Start station','End station','Bike number'],axis=1)
```

```python
target_data = [df]
target_map = {'Member':0,'Casual':1}
for data in target_data:
    data['Member type'] = data['Member type'].map(target_map)
```

```python
df.head()
```

```python
target = df['Member type']
df = df.drop(labels = 'Member type',axis=1)
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
x_train,x_test,y_train,y_test = train_test_split(df,target,test_size=0.2)
# model = GaussianNB()
# model.fit(x_train,y_train)
# pred = model.predict(x_test)
# print("Accuracy",accuracy_score(y_test,pred))
# print("Confusion matrix\n",confusion_matrix(y_test,pred))
```

```python
model1 = SVC(kernel='linear',C=1.5,probability=True)
model1.fit(x_train,y_train)
pred = model1.predict(x_test)
print("Accuracy",accuracy_score(y_test,pred))
print("Confusion matrix\n",confusion_matrix(y_test,pred))
```
